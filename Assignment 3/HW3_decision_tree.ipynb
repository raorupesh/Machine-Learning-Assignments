### Confusion matrix plot

```python
# Show confusion matrix
def plot_confusion_matrix(confusion_mat, cln):
    plt.imshow(confusion_mat, interpolation='nearest', cmap=plt.cm.gray)
    plt.title('Confusion matrix')
    plt.colorbar()
    tick_marks = np.arange(cln)
    plt.xticks(tick_marks, tick_marks)
    plt.yticks(tick_marks, tick_marks)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
y_true = [1, 0, 0, 2, 1, 0, 3, 3, 3]
y_pred = [1, 1, 0, 2, 1, 0, 1, 3, 3]
confusion_mat = confusion_matrix(y_true, y_pred)

print(confusion_mat)
plot_confusion_matrix(confusion_mat, 4)
```

# Decision Trees

### Load data

- For the following code, we use sklearn.datasets package for loading a dataset instead of reading a data file stored on a local machine. 

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

iris = load_iris()
#print(iris)
```

### Split the data to training and testing 

```python
X = iris.data[:, 2:] # petal length and width
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

## Training
### Learing using training data

- use Gini index measure 

*** Notes: you can also use gain information (entropy) measure by setting criterion="entropy" in the model

```python
from graphviz import Source
from sklearn.tree import export_graphviz

tree_clf = DecisionTreeClassifier(max_depth=2, criterion="gini", random_state=42)
tree_clf.fit(X_train, y_train)
```

## Testing
### Evaluating the model using testing data

```python
y_pred = tree_clf.predict(X_test)
```

# Visualization

## Confusion matrix

```python
# plot a confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)

print(confusion_mat)

plot_confusion_matrix(confusion_mat, 3)
```

### Model performance summary

```python
# Print classification report
target_names = iris.target_names

result_metrics = classification_report(y_test, y_pred, target_names=target_names)

print(result_metrics)
```

```python
# you can access each class's metrics from result_metrics
result_metrics_dict = classification_report(y_test, y_pred, target_names=target_names, output_dict=True)

print(result_metrics_dict['setosa']['precision'])
```

## k-Cross Validation

- using KFold function with freedom

```python
from sklearn.model_selection import KFold # import k-fold validation

kf = KFold(n_splits=3, random_state=None, shuffle=True) # Define the split - into 2 folds 

kf.get_n_splits(X) # returns the number of splitting iterations in the cross-validator

print(kf) 
```

### Applying k-Cross Validation

```python
tree_clf = DecisionTreeClassifier(max_depth=2, random_state=42)

for train_index, test_index in kf.split(X):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    tree_clf.fit(X_train, y_train)
    
    y_pred = tree_clf.predict(X_test)
    
    # Print classification report
    target_names = iris.target_names
    print(classification_report(y_test, y_pred, target_names=target_names))
```

# Decision Tree boundary Visualization

```python
## Example  This function is meant to be used for other data besides iris.  
```

```python
from matplotlib.colors import ListedColormap

def plot_decision_boundary(clf, X, y, axes=[0, 7.5, 0, 3], iris=True, legend=False, plot_training=True):
    x1s = np.linspace(axes[0], axes[1], 100)       # Return evenly spaced numbers over a specified interval.
    x2s = np.linspace(axes[2], axes[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s)
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)
    if not iris:
        custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
        plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    if plot_training:
        plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="Iris setosa")
        plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="Iris versicolor")
        plt.plot(X[:, 0][y==2], X[:, 1][y==2], "g^", label="Iris virginica")
        plt.axis(axes)
    if iris:
        plt.xlabel("Petal length", fontsize=14)
        plt.ylabel("Petal width", fontsize=14)
    else:
        plt.xlabel(r"$x_1$", fontsize=18)
        plt.ylabel(r"$x_2$", fontsize=18, rotation=0)
    if legend:
        plt.legend(loc="lower right", fontsize=14)

plt.figure(figsize=(8, 4))
plot_decision_boundary(tree_clf, X, y)
plt.plot([2.45, 2.45], [0, 3], "k-", linewidth=2)
plt.plot([2.45, 7.5], [1.75, 1.75], "k--", linewidth=2)
plt.plot([4.95, 4.95], [0, 1.75], "k:", linewidth=2)
plt.plot([4.85, 4.85], [1.75, 3], "k:", linewidth=2)
plt.text(1.40, 1.0, "Depth=0", fontsize=15)
plt.text(3.2, 1.80, "Depth=1", fontsize=13)
plt.text(4.05, 0.5, "(Depth=2)", fontsize=11)

save_fig("decision_tree_decision_boundaries_plot")
plt.show()
```

### Explain the relationship between the drawn decision boundaries and Gini index

##### The decision boundaries drawn using the entropy criterion are similar to those drawn using the Gini index. Both criteria aim to create the most homogeneous branches possible.

- The main difference is in the way they measure the impurity of a node.
- Gini index measures the impurity based on the probability of a randomly chosen element being incorrectly classified.
- Entropy measures the impurity based on the amount of information needed to describe the classification.
- In practice, both criteria often produce similar results, but entropy can be more computationally intensive.

```python
# Change the depth of the decision tree to 3, 4, and 5

depths = [3, 4, 5]
for depth in depths:
    tree_clf_entropy = DecisionTreeClassifier(max_depth=depth, criterion="entropy", random_state=42)
    tree_clf_entropy.fit(X_train, y_train)
    
    plt.figure(figsize=(8, 4))
    plot_decision_boundary(tree_clf_entropy, X, y, legend=False)
    plt.plot([2.45, 2.45], [0, 3], "k-", linewidth=2)
    plt.plot([2.45, 7.5], [1.75, 1.75], "k--", linewidth=2)
    plt.plot([4.95, 4.95], [0, 1.75], "k:", linewidth=2)
    plt.plot([4.85, 4.85], [1.75, 3], "k:", linewidth=2)
    plt.text(1.40, 1.0, "Depth=0", fontsize=15)
    plt.text(3.2, 1.80, "Depth=1", fontsize=13)
    plt.text(4.05, 0.5, f"(Depth={depth})", fontsize=11)
    
    save_fig(f"decision_tree_entropy_depth_{depth}_plot")
    plt.show()
```

```python
# Create the 5 fold cv
kf_5 = KFold(n_splits=5, random_state=42, shuffle=True)

# Validate output on the tree with depth 3
tree_clf_depth_3 = DecisionTreeClassifier(max_depth=3, criterion="entropy", random_state=42)
tree_clf_depth_3.fit(X_train, y_train)
scores_depth_3 = cross_val_score(tree_clf_depth_3, X, y, cv=kf_5)
print("5-Fold CV scores for depth 3:\n", scores_depth_3)
print("Mean 5-Fold CV score for depth 3:", scores_depth_3.mean())

# Create the 7 fold cv
kf_7 = KFold(n_splits=7, random_state=42, shuffle=True)

# Validate output on the tree with depth 4
tree_clf_depth_4 = DecisionTreeClassifier(max_depth=4, criterion="entropy", random_state=42)
tree_clf_depth_4.fit(X_train, y_train)
scores_depth_4 = cross_val_score(tree_clf_depth_4, X, y, cv=kf_7)
print("7-Fold CV scores for depth 4:\n", scores_depth_4)
print("Mean 7-Fold CV score for depth 4:", scores_depth_4.mean())

# Create the 10 fold cv
kf_10 = KFold(n_splits=10, random_state=42, shuffle=True)

# Validate output on the tree with depth 5
tree_clf_depth_5 = DecisionTreeClassifier(max_depth=5, criterion="entropy", random_state=42)
tree_clf_depth_5.fit(X_train, y_train)
scores_depth_5 = cross_val_score(tree_clf_depth_5, X, y, cv=kf_10)
print("10-Fold CV scores for depth 5:\n", scores_depth_5)
print("Mean 10-Fold CV score for depth 5:", scores_depth_5.mean())
```

### Explain your conclusions on increasing the depth and increasing the number of folds.

Increasing the depth of the decision tree generally improves the model's ability to capture the complexity of the data, leading to higher accuracy on the training set. However, it can also lead to overfitting, where the model performs well on the training data but poorly on unseen test data. This is because a deeper tree can capture noise and small fluctuations in the training data that do not generalize to new data.

Increasing the number of folds in cross-validation provides a more reliable estimate of the model's performance by reducing the variance associated with the random selection of the training and test sets. More folds mean that more data is used for training in each iteration, which can lead to a better understanding of the model's performance. However, it also increases the computational cost, as the model needs to be trained and evaluated more times.

In summary:
- Increasing the depth of the tree can improve training accuracy but may lead to overfitting.
- Increasing the number of folds in cross-validation provides a more reliable performance estimate but increases computational cost.
