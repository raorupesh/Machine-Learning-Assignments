````python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
# split the data into train and test parts
train_df, test_df = train_test_split(df,shuffle = True, test_size = 0.15, random_state=17)
# show descriptive charachteristics of the dataset such as min, max median
````

````python
# Scikit-Learn ≥0.20 is required
import sklearn
from sklearn import linear_model
# Select a linear model
import numpy as np
from sklearn.linear_model import LinearRegression

X= train_df['bmi']
y = train_df['charges']
X2= np.array(X).reshape(-1,1) # reshape(-1,1) makes the array horizontal, y is vertical
    
# Train the model (we use all data for training --> 
model = sklearn.linear_model.LinearRegression()
model.fit(X2, y)
r_sq = model.score(X2, y)
print(r_sq)
plt.scatter(X2, y, color='black', label='observed')

plt.plot(X2, model.predict(X2), label='fit', color='Green', linewidth=2)

plt.xlabel('bmi')
plt.ylabel('chages')
plt.title('Regression')
plt.legend(loc='best')

plt.show()
````

````python
# Make a prediction
X3= test_df['bmi']
 
X3= np.array(X3).reshape(-1,1)
# what is the charges for element 44 of the test bmi list. 
predicted_charge = model.predict([X3[44]])
print('bmi= {:.2f}   predicted charge = {:.2f}'.format(X3[44][0], predicted_charge[0]))
````

### Answer 2: Exploratory Data Analysis

````python
# Load the Real estate dataset
real_estate_df = pd.read_csv('Real estate.csv')

# Selected attributes (columns) of interest
selected_columns = ['X1 transaction date', 'X2 house age', 'X3 bus station', 'X4 number of stores', 'Y price sq feet']
real_estate_df = real_estate_df[selected_columns]


# Clean the data: Replace zeros with mean values
for column in real_estate_df.columns:
    real_estate_df[column] = real_estate_df[column].replace(0, real_estate_df[column].mean())

# Visualize correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = real_estate_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
````

### (a) The Mean, Median and Standar Deviation:

````python
# (a) The mean, median and standard deviation
mean_values = real_estate_df.mean()
median_values = real_estate_df.median()
std_values = real_estate_df.std()

print("Mean values:\n", mean_values)
print("\nMedian values:\n", median_values)
print("\nStandard Deviation values:\n", std_values)
````

### (b) Bloxplots for the attributes selected:

````python
# (b) Draw boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=real_estate_df, color='teal')
plt.title('Boxplot of Selected Attributes')
plt.show()
````

### (c) Histograms:

````python
# (c) Draw histograms
real_estate_df.hist(bins=50, figsize=(12, 10))
plt.suptitle('Histograms of Selected Attributes')
plt.show()
````

### (d) Scatterplots for the selected attributes:

````python
# (d) Draw scatter plots
sns.pairplot(real_estate_df)
plt.suptitle('Scatter Plots of Selected Attributes')
plt.show()
````

### (e) Simple linear regression function with a scatter plot:

````python
# (e) Simple linear regression function with a scatter plot
# Select features and target variable
X = real_estate_df[['X2 house age']]
y = real_estate_df['Y price sq feet']

# Train the model
model = LinearRegression()
model.fit(X, y)
r_sq = model.score(X, y)
print("Coefficient of determination (R^2):", r_sq)

# Plot the regression line
plt.scatter(X, y, color='blue', label='Observed')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('House Age')
plt.ylabel('House Price of Unit Area')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
````

````python
# Making a prediction
X3 = real_estate_df[['X2 house age']].values

# Predict the house price for element 100 of the house age list
predicted_price = model.predict([X3[100]])
print('House age = {:.2f}   Predicted price per sq feet = {:.2f}'.format(X3[100][0], predicted_price[0]))

# Evaluate the model performance
r_sq = model.score(X, y)
print("Coefficient of determination (R^2):", r_sq)
````
