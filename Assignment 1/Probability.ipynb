```python
import random
import matplotlib.pyplot as plt

def simulate_experiment():
    """
    Simulates a single experiment of tossing a coin until 3 heads are obtained.
    Returns the number of tosses needed to get 3 heads.
    """
    heads_count = 0
    tosses = 0
    while heads_count < 3:
        toss = random.choice(['H', 'T'])  # Simulate a coin toss
        tosses += 1
        if toss == 'H':
            heads_count += 1
    return tosses

def perform_experiments(num_experiments):
    """
    Performs the coin toss experiment multiple times and calculates the running average of tosses needed.
    Returns a list of running averages.
    """
    tosses_per_experiment = []
    running_averages = []

    for i in range(1, num_experiments + 1):
        tosses_needed = simulate_experiment()
        tosses_per_experiment.append(tosses_needed)
        running_averages.append(sum(tosses_per_experiment) / i)
    
    return running_averages

def plot_results(running_averages):
    """
    Plots the running averages of the number of tosses needed to get 3 heads.
    """
    plt.plot(range(1, len(running_averages) + 1), running_averages)
    plt.xlabel("Experiment Number")
    plt.ylabel("Average Number of Tosses to Get 3 Heads")
    plt.title("Average Number of Tosses to Get 3 Heads (Over 1000 Experiments)")
    plt.show()

# Perform the experiment 1000 times and plot the results
num_experiments = 1000
running_averages = perform_experiments(num_experiments)
plot_results(running_averages)
```
